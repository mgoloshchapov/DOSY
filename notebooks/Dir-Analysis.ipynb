```python
import pandas as pd
import os as os
```

```python
datadirs = {'Ri 128DOSY', 'R2-R4', 'R2-R5', 'R1-R2'}
n_min = 1
n_max = 5
method = 'BFGS'
reg = 0.01

for datadir in datadirs:
    results_ = {}
    dirname = str(pathlib.Path(home, 'data/'+datadir+'/'))
    for filename in os.listdir(dirname):
        if filename.endswith('.txt'):
            with open(os.path.join(dirname, filename), 'r') as f: # open in readonly mode
                expname = filename.split('.')[0]
                x, y = load_data(f.name, scale=1e6)
                params, m_aic, m_bic, cons_number = number_analysis(x, y, n_min=n_min, n_max=n_max, method=method, reg=reg)
                result = {}
                result['AIC'] = m_aic
                result['BIC'] = m_bic
                result['con'] = cons_number
                for i in range(n_min-1,n_max+1):
                    for j in range(n_min-1,2*i):
                        if j%2 == 0:
                            w = 'w'+str(j//2 + 1)+str(i)
                            result[w] = params[i-1][j]
                        else:
                            D = 'D'+str(j//2 + 1)+str(i)
                            result[D] = params[i-1][j]
        results_[expname] = result
    results = pd.DataFrame(results_).T
    #results.to_csv(datadir+'.csv')

    sheet_name = method + ' reg='+str(reg)
    excelname = '../results/'+datadir+'.xlsx'
    if os.path.exists(excelname):
        mode = 'a'
    else:
        mode = 'w'
    
    with pd.ExcelWriter(excelname, engine="openpyxl", mode = mode) as writer:  
        results.to_excel(writer, sheet_name=sheet_name)
    #results.to_excel(datadir+'.xlsx', sheet_name = sheet_name)
```
