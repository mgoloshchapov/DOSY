### R2-R4

```python
a = np.arange(1, 6)


W = np.round(a/(a+5.0), 2)
W = [(W[i], 1-W[i]) for i in range(len(W))]
W
```

```python
x, y, params_cf, params_da = dosy_analysis(home.joinpath("data/R4-R2", "R4-R2-5-1.txt"),
                                               method="curve_fit")

print(np.round(params_cf[1][0:2], 3))
print(np.round(params_cf[1][2:], 3))
```

```python
names = ["R4-R2-5-1", "R4-R2-5-2", "R4-R2-5-3", "R4-R2-5-4", "R4-R2-5-5"]

DR2 = []
DR4 = []

DR2_da = []
DR4_da = []

for i in range(1, 6):
    x, y, params_cf, params_da = dosy_analysis(home.joinpath("data/R4-R2", "R4-R2-5-{}.txt".format(i)),
                                               method="both")
    DR2.append(np.round(params_cf[1][0:2], 3))
    DR4.append(np.round(params_cf[1][2:], 3))
    DR2_da.append(np.round(params_da[1][0:2], 3))
    DR4_da.append(np.round(params_da[1][2:], 3))

res = pd.DataFrame({"R2-R4 weights" : W,
                    "R2, 1e-6, cf": DR2, "R4, 1e-6, cf": DR4,
                    "R2, 1e-6, da": DR2_da, "R4, 1e-6, da": DR4_da})
res.index = names
```
